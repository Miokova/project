```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np
from datetime import datetime, timedelta
from scipy.stats import mode
```

```python
sales = pd.read_csv("C:/Desktop/База данных - Продажи.csv")
customers = pd.read_csv("C:/Desktop/База данных - СПР_покупатели.csv")
products = pd.read_csv("C:/Desktop/База данных - СПР_продукты.csv")
managers = pd.read_csv("C:/Desktop/База данных - СПР_Менеджеры.csv")
shipping_cost = pd.read_csv("C:/Desktop/База данных - СПР_стоимость_доставки.csv")
order_date= pd.read_csv("C:/Desktop/База данных - СПР_даты_заказа.csv")
```

```python
order_date['Дата'] = pd.to_datetime(order_date['Дата'])
sales['Прибыль'] = pd.to_numeric(sales['Прибыль'], errors='coerce')
sales['Выручка'] = pd.to_numeric(sales['Выручка'].str.replace('\xa0' , ''), errors='coerce')
```
