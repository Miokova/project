 Аналитик данных отвечает за сбор, обработку и анализ данных по различным функциям бизнеса для принятия решений высшего руководства. Для работы аналитиком данных необходимы навыки аналитической мысли, навыки в области математики и строгость в работе с данными. Работа аналитиком данных подходит для тех, кто обладает высоким аналитическим складом ума и любопытством. Аналитик данных должен собирать, анализировать и интерпретировать данные, чтобы ответить на конкретные вопросы или решить проблемы в различных отраслях.
