В чем разница между операторами Delete и truncate?
Оператор DELETE удаляет строки из таблицы с возможностью использования условий WHERE для выбора конкретных строк, а оператор TRUNCATE удаляет все строки из таблицы без возможности использования условий WHERE и сбрасывает автоинкрементные значения.
В Python оператор DELETE удаляет отдельные строки из базы данных, а оператор TRUNCATE удаляет все строки из таблицы, но сохраняет структуру таблицы.

Какие соединения бывают в SQL?
INNER JOIN: возвращает строки, имеющие соответствующие значения в обеих таблицах.
LEFT JOIN (или LEFT OUTER JOIN): возвращает все строки из левой таблицы и соответствующие строки из правой таблицы.
RIGHT JOIN (или RIGHT OUTER JOIN): возвращает все строки из правой таблицы и соответствующие строки из левой таблицы.
FULL JOIN (или FULL OUTER JOIN): возвращает строки, имеющие соответствующие значения в любой из таблиц.
CROSS JOIN: возвращает декартово произведение строк из обеих таблиц.


В чем разница между типами данных CHAR и VARCHAR в SQL?
Тип данных CHAR в SQL хранит строку фиксированной длины, а тип данных VARCHAR хранит строку переменной длины. CHAR занимает фиксированное количество памяти, даже если строка короче, в то время как VARCHAR использует только столько памяти, сколько нужно для хранения строки.


Перечислить хотя бы 3 ограничения в SQL
PRIMARY KEY: Уникальный идентификатор записи в таблице.
FOREIGN KEY: Связь между двумя таблицами, обеспечивающая целостность данных.
UNIQUE: Гарантирует уникальность значений в столбце или наборе столбцов.
NOT NULL: Запрещает использование NULL значений в столбце.
CHECK: Позволяет определить условие, которое значения столбца должны удовлетворять.


Какую длину может иметь тип данных varchar2 в SQL, какую в PLSQL? В SQL тип данных VARCHAR2 может иметь длину до 4000 байт, а в PL/SQL - до 32767 байт.
