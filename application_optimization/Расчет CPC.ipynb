# 1. Расчёт мотивации для CPC-менеджеров.
Функция для расчета компенсации в зависимости от выполненных действий, реализации структуры тарификации с уровнями для менеджеров CPC.

```python
def calculate_payment(submission_count, additional_data_requests, cpc_agreements, cpc_refusals, reminders):
    base_payment = submission_count * 350  # базовая оплата за подачу заявок

    # Расчет дополнительной оплаты за выполненные действия
    additional_payment = additional_data_requests * <сумма за запросы доп. данных> + \
                        cpc_agreements * <сумма за согласия CPC> + \
                        cpc_refusals * <сумма за отказы CPC> + \
                        reminders * <сумма за напоминания>

    total_payment = base_payment + additional_payment  # общая сумма оплаты

    return total_payment
```


Чтобы решить проблему мотивации менеджеров CPC и обеспечить им стимул для выполнения всех необходимых действий по запросу, мы можем рассмотреть возможность пересмотра платежной системы. Одним из подходов могло бы стать внедрение многоуровневой структуры оплаты, при которой менеджеры получали бы дополнительную компенсацию за каждый этап процесса, который они выполняют по данному запросу.

Например, первоначальная отправка в CPC по-прежнему может быть компенсирована в размере 350 рублей за запрос. Однако последующим действиям, таким как обработка запросов дополнительных данных, получение соглашений о CPC, обработка отказов в CPC и отправка напоминаний, может быть присвоена определенная сумма компенсации. Такой многоуровневый подход мотивировал бы менеджеров доводить весь процесс до конца, поскольку каждый выполненный шаг добавлял бы дополнительную компенсацию.

Другой подход мог бы включать введение структуры бонусов, основанной на успешном завершении всего процесса. Например, менеджеры могли бы получать бонусы, если определенный процент их отправленных запросов приводил к успешным соглашениям с CPC. Это побудило бы менеджеров расставлять приоритеты в последующих действиях и повысило бы вероятность успешных результатов.

Кроме того, можно было бы внедрить систему стимулирования, основанную на результатах деятельности, при которой менеджеры получают долю от общей экономии или финансовой выгоды в результате успешных соглашений с CPC. Это привело бы их интересы в соответствие с финансовыми целями компании, мотивировало бы их добиваться положительных результатов и гарантировало бы, что их вознаграждение остается неизменным или немного выше, чем при предыдущей системе.

Эти изменения обеспечили бы адекватную мотивацию менеджеров CPC для выполнения всех необходимых действий и заключения соглашений с CPC, сохранив при этом общий уровень вознаграждения или немного повысив его.
