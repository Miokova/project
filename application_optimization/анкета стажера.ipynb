Почему в MSE в степени стоит квадрат, а не какая либо другая степень?

Среднеквадратичная ошибка (MSE) использует квадрат отклонений, потому что это позволяет учитывать не только величину ошибки, но и ее направление, так как отрицательные и положительные отклонения возводятся в квадрат, с тем чтобы избежать смешения положительных и отрицательных отклонений друг с другом.

Какому распределению соответствует задача бинарной классификации и почему?

Задача бинарной классификации соответствует распределению Бернулли, так как в данной задаче рассматривается ситуация, где исход может быть одним из двух возможных вариантов, а именно "успех" или "неудача".

Какому распределению соответствует функция ошибки MAE и почему?

Функция ошибки MAE соответствует распределению Лапласа из-за своего характера абсолютного отклонения и минимизации средних абсолютных ошибок, в отличие от квадратичного отклонения используемого в MSE
