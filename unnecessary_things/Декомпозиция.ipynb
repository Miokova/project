
<h1> Декомпозиция проекта "Ненужные вещи" <h1/>

### Загрузка и подготовка данных:
- Импорт библиотек и проверка данных на типы и отсутствие пропусков.

### Преобразование данных:
- Преобразование дат в формат datetime.
- Преобразование символов в строке в нижний регистр.

### Создание пользовательских профилей:
- Определение длительности сессии для событий и создание профилей пользователей.

### Обработка данных:
- Добавление столбцов с информацией о дне недели и времени суток.

### Анализ времени проведенного в приложении:
- Расчет времени жизни пользователя.

### Анализ удержания:
- Вычисление удержания пользователей и анализ изменений с течением времени.

### Анализ взаимодействия с пользователем:
- Определение событий контактов и анализ частоты действий.

### Анализ конверсии:
- Изучение конверсии в целевое действие - просмотр контактов.

### Визуализация и анализ данных:
- Создание графиков и тепловых карт для визуализации динамики событий,
  удержания и конверсии.

### Сегментация пользователей:
- Разделим пользователей на три группы на основе источника.  
- Проведем рассчет Retention rate в разрезе групп.
- Рассчитаем конверсию в целевое действие.

### Статистический анализ и проверка гипотез:
- Проведем Z-тест для определения статистически значимых различий по конверсии в просмотры контактов между двумя контрольными группами yandex и google. Сформулируем гипотезы: Нулевая гипотеза - конверсия в просмотры контактов одинакова среди пользователей из yandex и google; Альтернативная гипотеза - конверсия в просмотры контактов различается между пользователями из yandex и google.
    
    
- Проведение дополнительного Z-теста на разницу в долях между самым лучшим и самым худшим сегментом. Гипотезы: Нулевая гипотеза - конверсия в просмотры контактов одинакова между самым лучшим и самым худшим сегментом; Альтернативная гипотеза - конверсия в просмотры контактов различается между самым лучшим и самым худшим сегментом.
